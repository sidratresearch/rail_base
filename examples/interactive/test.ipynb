## Method

1. The interactive module reads all RAIL stages, and creates a wrapper function for each
    - That wrapper function runs `make_stage` and then `__call__`/`estimate`/`sample` etc.
2. Docstrings are read from a [stub file](../../src/rail/interactive/creation/degraders/unrec_bl_model.pyi)

```python
from rail.utils.interactive_utils import _initialize_interactive_module

_initialize_interactive_module("rail.interactive.creation.degraders", write_stubs=True)
```

## Limitations

- Docstrings must be correctly formatted (dev testing with numpydoc's validator)
- Stub files are to be distributed with rail-base, including functions from *all* RAIL packages
  - Because all RAIL packages share a namespace, we can't tell which package a stage came from - so users will see tab-complete options for functions that come from packages they don't have.
